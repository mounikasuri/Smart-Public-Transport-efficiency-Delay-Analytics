```
import pandas as pd
df = pd.read_csv('APSRTC_Transport_Data.csv')
print(df.head())
print(df.info())
```

```

print(df.isnull().sum())
df = df.fillna(df.mean(numeric_only=True))
df = df.dropna(subset=['bus_id','route','date'])
```

```

df = df.rename(columns={
    'bus_id': 'BusID',
    'route': 'Route',
    'date': 'Date',
})
df['Date'] = pd.to_datetime(df['Date'])
```

```

df.to_csv('APSRTC_Cleaned.csv', index=False)
```
